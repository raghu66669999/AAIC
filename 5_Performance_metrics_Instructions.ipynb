```python
# write your code here
import pandas as pd
import numpy as np

a=pd.read_csv('5_a.csv')
a['ypred']=[0 if i < 0.5 else 1 for i in a['proba']]
a.ypred.value_counts()
```

```python
a.head(10)
```

```python
print('confusion matrix:')
def confusion_mat(a):
    b1=[]
    for i in range(2):
        for j in range(2):
            b1.append(a.loc[(a['y']==j) & (a['ypred']==i)].shape[0])
    cm=np.array(b1).reshape(2,2)
    return cm
confusion_mat(a)
```

```python
print('#f1_score,#presion,#recall')

def f1_score(a):
    b=confusion_mat(a)
    pr=b[1][1]/(b[1][0]+b[1][1])
    re=b[1][1]/(b[0][1]+b[1][1])
    f1=2*pr*re/(pr+re)
    return f1,pr,re
f1_score(a)
```

```python
def tpr_fpr(a):
    b=confusion_mat(a)
    tpr=b[1][1]/(b[0][1]+b[1][1])
    fpr=b[1][0]/(b[0][0]+b[1][0])
    return tpr,fpr
tpr_fpr(a)
```

```python
def accuracy_score(a):
    b=confusion_mat(a)
    c=sum(b)
    acc=(b[0][0]+b[1][1])/sum(c)
    return acc
accuracy_score(a)
```

```python
print('#AUC')

def auc_score(a):
    z=sorted(list(a['proba']))
    tpr=[]
    fpr=[]
    for i in z:
        a['ypred']=[0 if k < i else 1 for k in a['proba']]
        tpr1,fpr1=tpr_fpr(a)
        tpr.append(tpr1)
        fpr.append(fpr1)
    return abs(np.trapz(tpr, fpr))
auc_score(a)
```

<ol>

<li> Compute MAPE: https://www.youtube.com/watch?v=ly6ztgIkUxk</li>

```python
d=pd.read_csv('5_d.csv')
m=list(map(lambda x,y:(x-y)**2,list(d['y']),list(d['pred'])))
mse=np.mean(m)
print('mse:',mse)
```

```python
#modified MAPE
m=list(map(lambda x,y:abs(x-y),list(d['y']),list(d['pred'])))
mape=sum(m)/sum(d['y'])
print('MAPE',mape)
```

```python
mean=np.mean(d['y'])
sstot=0.0
for i in range(len(d['y'])):
    sstot=sstot+(i-mean)**2
    
ssres= list(map(lambda x,y:(x-y)**2,list(d['y']),list(d['pred'])))
r2=(1-(sum(ssres)/sstot))
print('R2 error:',r2)
```
